## Model building

```python
from sklearn.model_selection import train_test_split
```

```python
train_set, test_set = train_test_split(log_air_passengers, train_size=0.8, shuffle=False)
```

```python
len(train_set), len(test_set)
```

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# grid search for finding the best models for our data
class model_evaluation:
    
    def __init__(self, data, trend_list, seasonal_list, damped_list, s_period):
        self.data = data
        self.trend_list = trend_list
        self.seasonal_list = seasonal_list
        self.damped_list = damped_list
        self.s_period = s_period
    
    def model_fitting(self, data, t_list, s_list, d_list, period):
        model = ExponentialSmoothing(data, trend=t_list, seasonal=s_list, damped=d_list, seasonal_periods=period)
        model_fit = model.fit()
        
        aic = model_fit.aic
        
        return aic
    
    def parameter_selection(self):
        
        best_aic, best_trend, best_seasonal, best_damp = float('inf'), None, None, None
        
        for t in self.trend_list:
            for s in self.seasonal_list:
                for d in self.damped_list:
                    for p in self.s_period:
                        try:
                            aic = self.model_fitting(self.data, t, s, d, p)
                            if aic < best_aic:
                                best_aic, best_trend, best_seasonal, best_damp, best_period = aic, t, s, d, p 
                            print(f'Trend: {t}, Seasonal: {s}, Damped: {d}, Seasonal Period: {p}, AIC: {aic}')
                        except:
                            continue
        
        print()
        print('Best parameters and AIC score:')
        print(f'Trend: {best_trend}, Seasonal: {best_seasonal}, Damped: {best_damp}, Seasonal Period: {best_period}, AIC: {best_aic}')
        
```

## Model evaluation and predictions

```python
model = ExponentialSmoothing(train_set, trend='add', seasonal='add', damped=False, seasonal_periods=12)
model_fit = model.fit()
```

```python
predicted_train_set = model_fit.fittedvalues
```

```python
predicted_train_set = np.exp(predicted_train_set)
```

```python
predicted_train_set.head()
```

```python
actual_train_set = np.exp(train_set)
```

```python
actual_test_set = np.exp(test_set)
```

```python
predicted_test_set = model_fit.predict(start=116, end=144)
```

```python
predicted_test_set = np.exp(predicted_test_set)
```

```python
plt.plot(air_passengers, label='actual')
plt.plot(predicted_train_set, color='red', linestyle='-.', label='training prediction')
plt.plot(predicted_test_set, color='green', linestyle='-', label='test prediction')
plt.xlabel('Years')
plt.ylabel('#thousand air passengers')
plt.title('Actual versus Predicted values of International Air Passengers')
plt.legend()
plt.tight_layout()
```

```python
plt.plot(actual_test_set, linestyle='-.', label='training prediction')
plt.plot(predicted_test_set, linestyle='-', label='test prediction')
plt.xlabel('Years')
plt.ylabel('#thousand air passengers')
plt.title('Actual vs Predicted on test set', fontsize=14)
plt.legend()
plt.tight_layout()
```

## Results

#### We will evaluate the statistical models and ANNs on the basis on Mean Absolute Error (MAE) and Mean Squared Error (MSE)

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
train_mae = mean_absolute_error(actual_train_set, predicted_train_set)
train_rmse = np.sqrt(mean_squared_error(actual_train_set, predicted_train_set))
train_nrmse = train_rmse/np.std(actual_train_set)

test_mae = mean_absolute_error(actual_test_set, predicted_test_set)
test_rmse = np.sqrt(mean_squared_error(actual_test_set, predicted_test_set))
test_nrmse = test_rmse/np.std(actual_test_set)

print(f'Training NRMSE: {train_nrmse}')
print(f'Training MAE: {train_mae}')

print(f'Test NRMSE: {test_nrmse}')
print(f'Test MAE: {test_mae}')
```
