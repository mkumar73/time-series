# Air Passenger forecasting using Dynamic harmonic regression

```r
library(readr)
library(dplyr)
library(ggplot2)
library(gridExtra)
library(xts)
library(zoo)
library(fpp2)
library(forecast)
library(Metrics)
```

```r
options(repr.plot.width=8, repr.plot.height=4)
```

## Data import and preparation

```r
data <- read_csv('../data/air-passenger/AirPassengers.csv')
head(data)

# change the name of column
names(data) <- c('Month', 'passengers')
print(head(data))
print(tail(data))


print(paste0('#samples', nrow(data)))
```

```r
# convert to time series object
# log transformed data
passenger <-
    ts(
        log(data[-1]),
        frequency = 12,
        start = c(1949, 1),
        end = c(1960, 12)
    )

head(passenger)
tail(passenger)
nrow(passenger)
```

```r
autoplot(passenger) + ylab('#passengers') + ggtitle('Air passenger data')
```

```r
## training and test split
train <- head(passenger, 115)
test <- tail(passenger, 29)

autoplot(train, series = 'train') + autolayer(test, series = 'test') +
    ylab('#passengers')
```

## ARMA and Fourier model

```r
## ARIMA model from SARIMA forecast sheet
# Drift included

model <- Arima(train, order = c(0, 1, 1), seasonal = list(order = c(1, 0, 1), period = 12),
               xreg = fourier(train, K = 6), lambda = 0, include.drift=TRUE)
frcst_value <- forecast(model, xreg=fourier(train, K=6, h=29))
```

```r
options(repr.plot.width=10, repr.plot.height=5)
```

```r
autoplot(exp(passenger), series = 'actual data') +
    autolayer(exp(fitted(model)), series = 'train prediction') +
    autolayer(exp(frcst_value$mean), series = 'test prediction') +
    xlab('Year') +
    ylab('#passengers') + 
    ggtitle('Acual v/s forecast result')
```

```r
summary(model)
```

## Model accuracy

```r
# train_rmse <- sqrt(mse(exp(train), train_pred))
# train_mae <- mae(exp(train), train_pred)

# test_rmse <- sqrt(mse(exp(test), test_pred))
# test_mae <- mae(exp(test), test_pred)
```

```r
train_pred <- exp(fitted(model))
test_pred <- exp(frcst_value$mean)

train_rmse <- sqrt(mse(exp(train), train_pred))
train_mae <- mae(exp(train), train_pred)
train_nrmse <- train_rmse/sd(exp(train))

test_rmse <- sqrt(mse(exp(test), test_pred))
test_mae <- mae(exp(test), test_pred)
test_nrmse <- test_rmse/sd(exp(test))


print(paste0('Training NRMSE :', round(train_nrmse, 3)))
print(paste0('Training MAE :', round(train_mae, 3)))

print(paste0('Test NRMSE :', round(test_nrmse, 3)))
print(paste0('Test MAE :', round(test_mae, 3)))
```

```r
autoplot(exp(test), series = 'test') +
    autolayer(test_pred, series = 'test prediction') +
    xlab('Year') +
    ylab('kWatt') +
    ggtitle('Actual vs predicted on test set using Arima and fourier series')
```
