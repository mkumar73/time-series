# Weather Forecasting using MPI data

```python
import pandas as pd
import numpy as np

import matplotlib.pylab as plt
%matplotlib inline

from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

%load_ext autoreload
%autoreload 2
%config InlineBackend.figure_format='retina'

import seaborn as sns
sns.set(style='ticks', context='poster')

plt.style.use('ggplot')
```

```python
data = pd.read_csv('data/mpi_weather_012018.csv', usecols=[0, 2], index_col=0, infer_datetime_format=True, parse_dates = [0])
```

```python
data.head()
```

```python
data = data.rename(index=str, columns={'Date Time': 'time', 'T (degC)': 'temperature'})
```

```python
fig, ax = plt.subplots()

data['temperature'].plot(ax=ax)
ax.set_title('Temperature from Jan-18 to June-18')
ax.set_xlabel('time period')
ax.set_ylabel('temperature (degree C)')
ax.xaxis.set_ticks_position('bottom')
# ax.set_xticks((np.arange(6)))
# ax.set_xticklabels(np.arange(6), ('Jan', 'Feb', 'March', 'Apr', 'May', 'June', 'July'))
fig.tight_layout()
```
